```python
df['Memory'] = df['Memory'].str.replace(r'\.','',regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0]
df["first"]=df["first"].str.strip()

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '')

df.loc[:,"second"]=df['second'].fillna("0")

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '',regex=True)
df['first'] = df['first'].str.replace(r'\D', '',regex=True)

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
import gradio as gr
import numpy as np
import unicodedata
def clean_unicode(text):
    if pd.isnull(text):
        return ''
    if isinstance(text,str):
        try:
            return unicodedata.normalize('NFKD',text).encode('acsii','ignore').decode('ascii')
        except:
            return text
        return str(text)
    
df.rename(columns={
    'TypeName':'Type',
    'Cpu_Brand':'CPU',
    'Gpu_Brand':'GPU',
    'os':'OS'
},inplace=True)

for col in ['Company','Type','CPU','GPU','OS']:
    if col in df.columns:
        df[col]=df[col].apply(clean_unicode)

def predict_price(company,type,ram,weight,touchscreen,ips,screen_size,resolution,cpu,hdd,ssd,gpu,os):
    try:
        company=clean_unicode(company)
        type=clean_unicode(type)
        cpu=clean_unicode(cpu)
        gpu=clean_unicode(gpu)
        os=clean_unicode(os)
        resolution=clean_unicode(resolution)
        if 'x' not in resolution.lower():
            return 'Resolution must be in the format 1920x1080...'
        x_res,y_res=map(int,resolution.lower().split('x'))
        
        ppi=((x_res**2+y_res**2)**0.5)/float(screen_size)
        touchscreen=1 if touchscreen=='Yes' else 0
        ips =1 if ips=='Yes' else 0
        query =np.array([company,type,ram,weight,touchscreen,ips,ppi,cpu,hdd,ssd,gpu,os])
        query=query.reshape(1,12)
        prediction=int(np.exp(pipe.predict(query)[0]))
        
        return f"Estimate Price:{prediction}"
    except Exception as e:
        return f"Error:{str(e)}"
```
